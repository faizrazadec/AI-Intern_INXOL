# Gender Classification Using DNN

### Problem Statement

In a modern and technologically advanced world, an intelligence agency is tasked with monitoring public spaces and gathering information for various purposes, including security and threat assessment. As part of their efforts, the agency is collecting vast amounts of data, including facial images of individuals, from surveillance cameras, social media, and other sources.


One specific task they are working on is to develop an AI-powered gender classification model. This model will be used to automatically classify the gender of individuals in the collected facial images. The agency believes that this information could provide valuable insights into demographic patterns and potentially aid in identifying persons of interest more efficiently.

### Data Collection
Data is used from the 

```python
folder_paths = [
    'C:/Users/faizr/Downloads/New folder/Female',
    'C:/Users/faizr/Downloads/New folder/Male'
]
```

### Reading the data and preprocessing

```python
#took 40  minutes
#just required libraries
import cv2 as cv
import os
import numpy as np
import seaborn as sns
import time

start_time = time.time()
#dataset an empty list is created to store the images in array form
dataset = []

#iterating through the folders
for i in folder_paths:
    folder_name = os.path.basename(i)  #storing the folder name in "folder_name"
    
    #iterating trough the files inside the folders
    for file_name in os.listdir(i):
        image_path = os.path.join(i, file_name)  #storing the images paths name in "image_path"
        
        #"os.path.isfile" checking if the "image_path" is present then read the image
        if os.path.isfile(image_path):
            image = cv.imread(image_path, cv.IMREAD_GRAYSCALE)    #reading the image and then turing it into the grayscale
            
            if image is not None:   #if image is successfully readed, 
                # Load the Haarcascades Classifier
                haar_face_cascade = cv.CascadeClassifier('C:/Users/faizr/Downloads/haarcascade_frontalface_alt.xml')

                # Detect faces
                faces = haar_face_cascade.detectMultiScale(image)

                # Print the number of faces found
                #print('Faces found:', len(faces))
                
                #"face_matrics" is an empty list to store the pixels of just the face that we detected
                face_matrices = []
                
                #extracting the faces
                for (x, y, w, h) in faces:
                    face_matrix = image[y:y+h, x:x+w]
                    face_matrices.append(face_matrix)
                
                # if the image array is successfully appended to the list (face is detected)
                if len(face_matrices) > 0:
                    
                    #fetching just the first image, if there are more faces in the image
                    face = face_matrices[0]
                    resize = cv.resize(face, (150, 150))    #resizing the face so all the image can be of same size
                    flattened_face = np.array(resize).flatten().tolist()  #turning it into one-D array then to list

                    #print("Number of columns:", len(flattened_face))
                    dataset.append(flattened_face + [folder_name])   #appending the arraies of faces to the "dataset" that we created above
                else:
                    print("No faces found in the image:", image_path)    #if face couldn't be detected
Reading_time = time.time() - start_time
Reading_time = Reading_time/60
print(f"Reading Time: {Reading_time:.4f} minutes")
```

### Making the DataFrame

```python
import pandas as pd
data = pd.DataFrame(dataset, columns=[f'pixel_{i+1}' for i in range(len(flattened_face))] + ['label'])
data
```

Preprocessing Process
Randomizing the Labels, e.g from the sequance of Male and Female to "male,male, female, male, female" etc
it'll help our model to reduce the biasness

```python
randomized_indices = np.random.permutation(data.shape[0])
randomized_data = data.iloc[randomized_indices]
randomized_data
```

### Features and Target
Splitting the data into features and target

```python
x = randomized_data.drop('label', axis = 1) /255
y = randomized_data['label']
y.value_counts()    #counting the classes in the label
```

DNN works on Numerical data, if the data is categorical or non-numerical, then we have to encode it using encoders

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y_encoded = encoder.fit_transform(y)
y_encoded
```

### Dataset Splitting
Divide the dataset into training, validation, and test sets. The training set is used to train the model, the validation set helps tune hyperparameters, and the test set evaluates the final model's performance.

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y_encoded, test_size = 0.2, random_state = 42)
```

### Model Selection
Here we imported the Sequentional Model from the keras.models
As we have to use the DNN, DNN is the Dense Neural Network, inn which the layers are densely connected with all others layers, first is the input layer, last is the output layer, and the rest of layers are hidden layers that plays the magic in DNN or NN

```python
from keras.models import Sequential
from keras.layers import Dense
model = Sequential()

model.add(Dense(128, activation = 'relu', input_shape=(150*150,)))
model.add(Dense(64, activation = 'relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(len(encoder.classes_), activation = 'softmax'))
model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
model.summary()
```

Here the target is converting into tensors as DNN takes the tensors as input, not the array or list that is faltten

```python
from keras.utils import to_categorical
y_train_encoded = to_categorical(y_train)
y_test_encoded = to_categorical(y_test)
x_train.shape, y_train_encoded.shape
```

At the end, as usual, we fit the model!

```python
start_time = time.time()
model.fit(x_train, y_train_encoded, batch_size = 32, epochs =50, validation_split = 0.2)
Reading_time = time.time() - start_time
Reading_time = Reading_time/60
print(f"Reading Time: {Reading_time:.2f} minutes")
```

### Evaluating the Model

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
y_predict = model.predict(x_test)
y_predict_max = np.argmax(y_predict, axis = 1)

accuracy = accuracy_score(y_test, y_predict_max)*100
classi = classification_report(y_test, y_predict_max)
conmat = confusion_matrix(y_test, y_predict_max)

print(f"The accuracy of the model is: {accuracy: .2f}")
print(classi)

plt.figure(figsize=(6, 4))
sns.heatmap(conmat, annot=True, fmt=".1f", linewidth=1, cmap="crest")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.title(f"Confusion Matrix")
plt.show()
```
